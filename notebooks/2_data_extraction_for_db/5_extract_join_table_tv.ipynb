```python
import pandas as pd
import os
```

```python
tv = pd.read_csv(os.path.join("../datasets", "tv.csv"))
```

For tv  

TV join tables:

For TVShowCreator

```python
creators_df = pd.read_csv(os.path.join("../db_tables", "Creator_rows.csv"))
tv_creators_rows = []
```

```python
for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.created_by):
        continue

    creators = str(row.created_by).split(",")

    for creator in creators:
        creator = creator.strip()

        creator_id = creators_df.loc[creators_df["name"] == creator, 'id'].values[0]
        tv_creators_rows.append({
            "tvShowId": tv_id,
            "creatorId": creator_id
        })
```

```python
tv_creator_df = pd.DataFrame(tv_creators_rows)
tv_creator_df.to_csv(os.path.join('../datasets', 'tv_creator.csv'), index=False)
```

For TVShowLanguage, TVShowSpokenLanguage TVShowOriginalLanguage

```python
spoken_language_df = pd.read_csv(os.path.join("../db_tables", "SpokenLanguage_rows.csv"))
language_df = pd.read_csv(os.path.join("../db_tables", "Language_rows.csv"))

tv_sl_rows = []
tv_lang_rows = []
tv_ol_rows = []
```

```python
for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.spoken_languages):
        continue

    spoken_languages = str(row.spoken_languages).split(",")

    for spoken_language in spoken_languages:
        spoken_language = spoken_language.strip()

        spoken_language_id = spoken_language_df.loc[spoken_language_df["name"] == spoken_language, 'id'].values[0]
        tv_sl_rows.append({
            "tvShowId": tv_id,
            "languageId": spoken_language_id
        })

for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.languages):
        continue

    languages = str(row.languages).split(",")

    for language in languages:
        language_code = language.strip()
        tv_lang_rows.append({
            "tvShowId": tv_id,
            "languageCode": language_code
        })

for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.original_language):
        continue

    original_languages = str(row.original_language).split(",")

    for original_language in original_languages:
        original_language_code = original_language.strip()
        tv_ol_rows.append({
            "tvShowId": tv_id,
            "languageCode": original_language_code
        })
```

```python
tv_sl_df = pd.DataFrame(tv_sl_rows)
tv_lang_df = pd.DataFrame(tv_lang_rows)
tv_ol_df = pd.DataFrame(tv_ol_rows)

tv_sl_df.to_csv(os.path.join('../datasets', 'tv_spoken_language.csv'), index=False)
tv_lang_df.to_csv(os.path.join('../datasets', 'tv_language.csv'), index=False)
tv_ol_df.to_csv(os.path.join('../datasets', 'tv_original_language.csv'), index=False)
```

For TVShowProductionCompany, TVShowProductionCountry

```python
production_company_df = pd.read_csv(os.path.join("../db_tables", "ProductionCompany_rows.csv"))
production_country_df = pd.read_csv(os.path.join("../db_tables", "ProductionCountry_rows.csv"))

tv_production_company_rows = []
tv_production_country_rows = []
```

```python
for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.production_companies):
        continue

    production_companies = str(row.production_companies).split(",")

    for production_company in production_companies:
        production_company = production_company.strip()

        match = production_company_df.loc[production_company_df["name"].str.strip() == production_company, "id"]

        if match.empty:
            continue

        production_company_id = match.iloc[0]
        tv_production_company_rows.append({
            "tvShowId": tv_id,
            "companyId": production_company_id
        })
```

```python
for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.production_countries):
        continue

    production_countries = str(row.production_countries).split(",")

    for production_country in production_countries:
        production_country = production_country.strip()

        match = production_country_df.loc[production_country_df["name"].str.strip() == production_country, "id"]

        if match.empty:
            continue

        production_country_id = match.iloc[0]

        tv_production_country_rows.append({
            "tvShowId": tv_id,
            "countryId": production_country_id
        })
```

```python
tv_production_company_rows = pd.DataFrame(tv_production_company_rows)
tv_production_company_rows.to_csv(os.path.join('../datasets', 'tv_production_company.csv'), index=False)

tv_production_country_rows = pd.DataFrame(tv_production_country_rows)
tv_production_country_rows.to_csv(os.path.join('../datasets', 'tv_production_country.csv'), index=False)
```

For TVShowCast

```python
person_df = pd.read_csv(os.path.join("../db_tables", "Person_rows.csv"))

tv_cast_rows = []
```

```python
import ast

for row in tv.itertuples(index=False):
    tv_id = row.id

    if pd.isna(row.cast):
        continue

    try:
        cast_list = ast.literal_eval(row.cast) 
    except (ValueError, SyntaxError):
        print("error")
        continue

    for cast_dict in cast_list:
        match = person_df.loc[person_df["id"] == cast_dict["id"], "id"]

        if match.empty:
            continue

        tv_cast_rows.append({
            "tvId": tv_id,
            "personId": match.iloc[0],
            "character": cast_dict["character"],
            "castOrder": cast_dict["order"]
        })
```

```python
tv_cast_rows = pd.DataFrame(tv_cast_rows)
tv_cast_rows.to_csv(os.path.join('../datasets', 'tv_cast.csv'), index=False)
```
