```python
import pandas as pd
```

```python
movies_df = pd.read_csv("../datasets/movies.csv")
```

```python
movies = movies_df[[
    "id",
    "title",
    "original_title",
    "overview",
    "tagline",
    "status",
    "release_date",
    "runtime",
    "budget",
    "revenue",
    "adult",
    "homepage",
    "imdb_id",
    "popularity",
    "vote_average",
    "vote_count",
    "poster_path",
    "backdrop_path"
]].copy()

movies.rename(columns={
    "original_title": "originalTitle",
    "release_date": "releaseDate",
    "imdb_id": "imdbId",
    "vote_average": "voteAverage",
    "vote_count": "voteCount",
    "poster_path": "posterPath",
    "backdrop_path": "backdropPath"
}, inplace=True)
```

```python
movies["releaseDate"] = pd.to_datetime(movies["releaseDate"], errors="coerce")
movies["adult"] = movies["adult"].astype(bool)
```

```python
movies.to_csv("../datasets/movies_table.csv", index=False)
```
