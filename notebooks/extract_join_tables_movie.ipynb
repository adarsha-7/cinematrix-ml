```python
import pandas as pd
import os
```

```python
movies = pd.read_csv(os.path.join("../datasets", "movies.csv"))
```

For movies

Movie join tables:

  genres              MovieGenre[]  
  keywords            MovieKeyword[]  
  spokenLanguages     MovieSpokenLanguage[]  
  originalLanguages   MovieOriginalLanguage[]  
  productionCompanies MovieProductionCompany[]  
  productionCountries MovieProductionCountry[]  
  cast                MovieCast[]  
  director            MovieDirector[]  

For MovieKeyword

```python
keywords_df = pd.read_csv(os.path.join("../db_tables", "Keyword_rows.csv"))
movie_keyword_rows = []
```

```python
for row in movies.itertuples(index=False):
    movie_id = row.id

    if pd.isna(row.keywords):
        continue

    keywords = str(row.keywords).split(",")

    for keyword in keywords:
        keyword = keyword.strip()

        keyword_id = keywords_df.loc[keywords_df["name"] == keyword, 'id'].values[0]
        movie_keyword_rows.append({
            "movieId": movie_id,
            "keywordId": keyword_id
        })
```

```python
movie_keyword_df = pd.DataFrame(movie_keyword_rows)
movie_keyword_df.to_csv(os.path.join('../datasets', 'movie_keyword.csv'), index=False)
```

For MovieProductionCompany, MovieProductionCountry

```python
production_company_df = pd.read_csv(os.path.join("../db_tables", "ProductionCompany_rows.csv"))
production_country_df = pd.read_csv(os.path.join("../db_tables", "ProductionCountry_rows.csv"))

movie_production_company_rows = []
movie_production_country_rows = []
```

```python
for row in movies.itertuples(index=False):
    movie_id = row.id

    if pd.isna(row.production_companies):
        continue

    production_companies = str(row.production_companies).split(",")

    for production_company in production_companies:
        production_company = production_company.strip()

        match = production_company_df.loc[production_company_df["name"].str.strip() == production_company, "id"]

        if match.empty:
            continue

        production_company_id = match.iloc[0]
        movie_production_company_rows.append({
            "movieId": movie_id,
            "companyId": production_company_id
        })
```

```python
for row in movies.itertuples(index=False):
    movie_id = row.id

    if pd.isna(row.production_countries):
        continue

    production_countries = str(row.production_countries).split(",")

    for production_country in production_countries:
        production_country = production_country.strip()

        match = production_country_df.loc[production_country_df["name"].str.strip() == production_country, "id"]

        if match.empty:
            continue

        production_country_id = match.iloc[0]

        movie_production_country_rows.append({
            "movieId": movie_id,
            "countryId": production_country_id
        })
```

```python
movie_production_company_rows = pd.DataFrame(movie_production_company_rows)
movie_production_company_rows.to_csv(os.path.join('../datasets', 'movie_production_company.csv'), index=False)

movie_production_country_rows = pd.DataFrame(movie_production_country_rows)
movie_production_country_rows.to_csv(os.path.join('../datasets', 'movie_production_country.csv'), index=False)
```

For MovieDirector, MovieCast

```python
person_df = pd.read_csv(os.path.join("../db_tables", "Person_rows.csv"))

movie_cast_rows = []
movie_director_rows = []
```

```python
import ast

for row in movies.itertuples(index=False):
    movie_id = row.id

    if pd.isna(row.cast):
        continue

    try:
        cast_list = ast.literal_eval(row.cast) 
    except (ValueError, SyntaxError):
        print("error")
        continue

    for cast_dict in cast_list:
        match = person_df.loc[person_df["id"] == cast_dict["id"], "id"]

        if match.empty:
            continue

        movie_cast_rows.append({
            "movieId": movie_id,
            "personId": match.iloc[0],
            "castId": cast_dict["cast_id"],
            "character": cast_dict["character"],
            "castOrder": cast_dict["order"]
        })
```

```python
for row in movies.itertuples(index=False):
    movie_id = row.id

    if pd.isna(row.director):
        continue

    try:
        director = ast.literal_eval(row.director) 
    except (ValueError, SyntaxError):
        print("error")
        continue

    match = person_df.loc[person_df["id"] == director["id"], "id"]

    if match.empty: 
        continue

    movie_director_rows.append({
            "movieId": movie_id,
            "personId": match.iloc[0]
        })
```

```python
movie_cast_rows = pd.DataFrame(movie_cast_rows)
movie_cast_rows.to_csv(os.path.join('../datasets', 'movie_cast.csv'), index=False)

movie_director_rows = pd.DataFrame(movie_director_rows)
movie_director_rows.to_csv(os.path.join('../datasets', 'movie_director.csv'), index=False)
```
