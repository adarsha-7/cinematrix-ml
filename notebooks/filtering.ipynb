```python
import pandas as pd
import os
```

```python
datasets_path = "../datasets"

movie_filename = "TMDB_movie_dataset.csv"
tv_filename = "TMDB_tv_dataset.csv"
```

```python
movies_df = pd.read_csv(os.path.join(datasets_path, movie_filename))

print(movies_df.shape)
```

```python
print(movies_df[movies_df['vote_count'] <= 5].shape)
```

The original raw dataset has 1.326m entries. We will filter out 1.215m entries which will probably be not necessary for us indicated by
the fact that they have less than or equal to just 5 votes.
This will leave us with about 111k entries which is a sufficent and good size for our database.

```python
top_movies = movies_df[movies_df['vote_count'] > 5]
print(top_movies.shape)
```

```python
top_movies_filename = "top_movies.csv"
top_movies.to_csv(os.path.join(datasets_path, top_movies_filename))

top_movies_df = pd.read_csv(os.path.join(datasets_path, top_movies_filename))
print(top_movies_df.shape)
```

Doing the same for series dataset

```python
tv_df = pd.read_csv(os.path.join(datasets_path, tv_filename))

print(tv_df.shape)
```

```python
print(tv_df[tv_df['vote_count'] <= 5].shape)
```

```python
top_tv = tv_df[tv_df['vote_count'] > 5]
print(top_tv.shape)
```

```python
top_tv_filename = "top_tv.csv"
top_tv.to_csv(os.path.join(datasets_path, top_tv_filename))

top_tv_df = pd.read_csv(os.path.join(datasets_path, top_tv_filename))
print(top_tv_df.shape)
```

```python
print(movies_df.memory_usage(deep=True).sum() / (1024**2))
print(top_movies_df.memory_usage(deep=True).sum() / (1024**2))
print(tv_df.memory_usage(deep=True).sum() / (1024**2))
print(top_tv_df.memory_usage(deep=True).sum() / (1024**2))
```
