```python
import pandas as pd
import os
```

```python
movies = pd.read_csv(os.path.join('../datasets', 'movies.csv'))
tv = pd.read_csv(os.path.join('../datasets', 'tv.csv'))
```

For keywords

```python
keywords_column = movies["keywords"].dropna()
keywords = [keyword.strip() for row in keywords_column for keyword in row.split(',')]
keywords_unique = list(set(keywords))
keywords_df = pd.DataFrame(keywords_unique, columns=['name'])
keywords_df
```

```python
keywords_df.to_csv('../datasets/keywords.csv', index=False)
```

For genre

```python
genre_column_movies = movies['genres'].dropna()
genre_column_tv = tv['genres'].dropna()
genre_column = pd.concat([genre_column_movies, genre_column_tv], axis=0)
genres = [genre.strip() for row in genre_column for genre in row.split(',')]
genres_unique = list(set(genres))
genres_df = pd.DataFrame(genres_unique, columns=['name'])
genres_df
```

```python
genres_df.to_csv('../datasets/genres.csv', index=False)
```

For production company

```python
production_company_column_movies = movies['production_companies'].dropna()
production_company_column_tv = tv['production_companies'].dropna()
production_company_column = pd.concat([production_company_column_movies, production_company_column_tv], axis=0)
production_companies = [production_company.strip() for row in production_company_column for production_company in row.split(',')]
production_companies_unique = list(set(production_companies))
production_companies_df = pd.DataFrame(production_companies_unique, columns=['name'])
```

```python
production_companies_df.to_csv('../datasets/production_companies.csv', index=False)
```

For original language

```python
original_language_column_movies = movies['original_language'].dropna()
original_language_column_tv = tv['original_language'].dropna()
language_column_tv = tv['languages'].dropna()
language_column = pd.concat([original_language_column_movies, original_language_column_tv, language_column_tv], axis=0)
languages = [language.strip() for row in language_column for language in row.split(',')]
languages_unique = list(set(languages))
languages_df = pd.DataFrame(languages_unique, columns=['code'])
```

```python
languages_df.to_csv('../datasets/languages.csv', index=False)
```

Note: the file languages.csv is manually changed to add column called "name" which is the language for given code

For person

```python
actors = pd.read_csv(os.path.join('../datasets', 'actors.csv'))
directors = pd.read_csv(os.path.join('../datasets', 'directors.csv'))

person_column = pd.concat([actors, directors], axis=0)
person_column.drop_duplicates(subset='id', keep='first', inplace=True)
```
